```python
import re
import numpy as np
```

```python
data = open('shakespeare.txt', 'r').read()
chars = list(set(data))
```

```python
chars = sorted(chars)
char_to_ix = { ch:i for i,ch in enumerate(chars) }
ix_to_char = { i:ch for i,ch in enumerate(chars) }
ix_to_char
```

```python
characters = []
with open("shakespeare.txt") as text:
    for line in text:
        for ch in line:#.lower():
            characters.append(ch)
characters
```

```python
code = []
for ch in characters:
    code.append( char_to_ix[ch] )
code
```

```python
np.savetxt("code.txt", code, fmt="%04d")
```

```python
lyrics_ch = []
lyrics_df = pandas.read_csv('songdata.csv')
for song in lyrics_df['text']:
    for ch in song:
        lyrics_ch.append(ch)
```

```python
lyrics_code = []
for ch in lyrics_ch:
    if ch in char_to_ix.keys():
        lyrics_code.append(char_to_ix[ch])
    else:
        lyrics_code.append(1)
lyrics_code
```

```python
np.savetxt("lyrics_code.txt", lyrics_code[:200000], fmt="%04d")
```
